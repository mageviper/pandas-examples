```python
import pandas as pd
```

```python
#load customer data
customer_transactions = pd.read_csv("data/customer-transactions.csv")
customer_transactions

```

```python
last_purchase_date = pd.to_datetime(customer_transactions['transaction_date'])
grouped = customer_transactions.groupby('customer_id').agg({'transaction_date': 'max',})
grouped['recency'] = (pd.Timestamp.today() - pd.to_datetime(grouped['transaction_date']))
grouped['recency'] = grouped['recency'].dt.days
grouped

```

```python

frequency_count = customer_transactions['customer_id'].value_counts()
customer_transactions['frequency'] = customer_transactions['customer_id'].map(frequency_count)
customer_transactions
```

```python
customer_rfm = customer_transactions.groupby('customer_id').agg({
    'transaction_date': 'max',  # do obliczenia recency
    'transaction_amount': 'sum',  # monetary
    'customer_id': 'count'  # frequency
})
customer_rfm.columns = ['recency', 'monetary', 'frequency']
customer_rfm['recency'] = (pd.Timestamp.today() - pd.to_datetime(customer_rfm['recency'])).dt.days
customer_rfm
```
